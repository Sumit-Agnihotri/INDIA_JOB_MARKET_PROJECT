```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('india_job_market_dataset.csv')
dataset
```

# What are the most in-demand job roles?

```python
sns.countplot(x='Job Title',data=dataset, hue="Job Title")
plt.xticks(rotation=90)
plt.xlabel('Job Title', fontsize=20, color='Green')
plt.ylabel('Count', fontsize=20, color='blue')
plt.title('Job Title vs Count', fontsize=20, color='red')
plt.show()
```

# Which companies are posting most job openings?

```python
sns.countplot(x="Company Name",data=dataset, hue="Company Name")
plt.xticks(rotation=90)
plt.xlabel("Company Name", fontsize=20, color="Green")
plt.ylabel("Count", fontsize=20, color="blue")
plt.title("Company Name vs Count", fontsize=20, color="red")
plt.show()
```

# What is the distribution of Job Types?

```python
gb=dataset.groupby("Job Type").agg({"Job Type":"count"})
colors=["Red","Yellow","Green","Blue"]
plt.pie(gb["Job Type"], labels=gb.index, autopct="%1.1f%%", colors=colors)
print(gb)
plt.show()
```

# What is the Distribution of Salary ranges based on Job Ranges?

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Job Title",data=dataset, hue="Salary Range")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
